```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import os
```

```python
path = r"----"
```

```python
os.listdir(path)
```

```python
# Create list with files to merge based on name convention
file_list = [path + f for f in os.listdir(path)]

# Empty list to include the content of each file converted to df
csv_list = []
 
# Read each (sorted) file in file_list, converts it to pandas df and appends it to the csv_list
for file in sorted(file_list):
    csv_list.append(pd.read_csv(file).assign(File_Name = os.path.basename(file)))

# Merges single pandas df into a single df, index is refreshed 
csv_merged = pd.concat(csv_list, ignore_index=True)

# Single df is saved to the path in CSV format, without index column
csv_merged.to_csv(path + 'registro-obsequios_full.csv', index=False)
```

```python
from difflib import SequenceMatcher

unique_names = df['Funcionario'].unique()

threshold = 0.75
candidate_similar_names = []
for i, name1 in enumerate(unique_names[:-1]):
    for name2 in unique_names[i+1:]:
        similarity = SequenceMatcher(None, name1, name2).ratio()
        if similarity > threshold:
            candidate_similar_names.append((name1, name2, similarity))
            
candidate_similar_names
```
